## 3. Capturing NAN values with a new feature  

1. This technique works best when data is not missing completely at random. In this, we add a new column in the dataset and replace all NaN values with 1 or 0.

2. By doing this we get easy visualization of null values i.e, we can easily track the null values

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Harsh Jain\Desktop\HARSH JAIN\3. EVERYTHING RELATED TO DATA\DATA SCIENCE\DATASETS\titanic train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python
#Here we'll create a new feature Age_Nan
df['Age_NAN']=np.where(df['Age'].isnull(),1,0)
```
